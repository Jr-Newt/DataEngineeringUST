```python
import numpy as np
```

#### 13. Merge

```python
df_series1 =pd.Series( {
    'USA': 21,
    'China': 14,
    'Japan': 5,
    'Germany': 4,
    'India': 2.9,
    'UK': 2.8,
    'France': 2.7,
    'Brazil': 1.4,
    'Italy': 2.1,
    'Canada': 1.6
}, name  = 'GDP_Per_Capita')
```

```python
df_series2 =pd.Series( {
    'USA': 98,
    'China': 44,
    'Japan': 66,
    'Germany': 54,
    'India': 29,
    'UK': 78,
    'France': 47,
    'Brazil': 24,
    'Italy': 91,
    'Canada': 26
}, name  = 'Population')
```

```python
df1 = pd.DataFrame(df_series1).reset_index()
```

```python
# renaming the existing column
df1.columns = ['Country','GDP']
```

```python
# renaming existing column
df2 = pd.DataFrame(df_series2).reset_index()
df2.columns = ['Country','Population']
```

```python
pd.merge(df1, df2 , on = 'Country', how = 'inner')
```

```python
pd.merge(df1, df2 , on = 'Country', how = 'outer')
```

```python
pd.merge(df1, df2 , on = 'Country', how = 'left')
```

```python
pd.merge(df1, df2 , on = 'Country', how = 'right')
```
