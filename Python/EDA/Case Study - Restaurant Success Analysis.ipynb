```python
from geopy.geocoders import Nominatim
import folium
```

```python
locations = pd.DataFrame({"Name":zomato['location'].unique()})
locations['Name'] = locations['Name'].apply(lambda x : "Bangalore " + str(x))
locations
```

```python
from geopy.geocoders import Nominatim
import numpy as np
import time

# Initialize the geocoder
geo_loc = Nominatim(user_agent="app")

lat_long = []
max_retries = 5
retry_delay = 5  # seconds

for location_name in locations['Name']:
    retries = 0
    while retries < max_retries:
        try:
            location = geo_loc.geocode(location_name, timeout=10)
            if location is None:
                lat_long.append(np.nan)
            else:
                geo = (location.latitude, location.longitude)
                lat_long.append(geo)
            break  # Exit loop if request was successful
        except Exception as e:
            print(f"Error for location '{location_name}': {e}")
            retries += 1
            if retries < max_retries:
                time.sleep(retry_delay)  # Wait before retrying
            else:
                lat_long.append(np.nan)
                print(f"Max retries reached for location '{location_name}'. Appending NaN.")
```

```python
locations['geo_loc'] = lat_long
locations.head()
```
